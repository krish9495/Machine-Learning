```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
X,y=make_classification(n_features=5,n_informative=5,n_redundant=0,n_clusters_per_class=1)
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
```

```python
df.shape
```

```python
df.head()
```

```python
bag=BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,0:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,6))
plot_tree(bag.estimators_[0])
plt.show()
```

# All we have is X[1] and X[0]

### only these 2 got selected

#### like already 2 column randomly selected and whole tree will be made from them

# Now will see with RandomForestClassifier

```python
rf=RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,0:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,6))
plot_tree(rf.estimators_[0])
plt.show()
```

# All the column are randomly selected at the node level
