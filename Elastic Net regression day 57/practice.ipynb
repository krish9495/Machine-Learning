```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"D:\100 Days machine learning\Elastic Net regression day 57\diabetesData.csv")
```

```python
df.head()
```

```python
X=df.iloc[:,0:8]
```

```python
y=df.iloc[:,-1]
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse_output=False)
```

```python
y_new=y.values.reshape(-1,1)
```

```python
y_new
```

```python
y_new1=ohe.fit_transform(y_new)
```

```python
y_new1
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y_new1,test_size=0.2,random_state=2)
```

```python
#Linear regression
reg=LinearRegression()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Ridge 
reg = Ridge(alpha=0.1)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Lasso
reg = Lasso(alpha=0.01)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# ElasticNet
reg = ElasticNet(alpha=0.005,l1_ratio=0.9)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```
