```python
from sklearn.datasets import make_regression
```

```python
import numpy as np 
import pandas as pd
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
#applying OLS first
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
#this is m
```

```python
reg.intercept_
#this is b0
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

##  Lets apply Gradient Descent assuming slope is constant m = 78.35
## and let's assume the starting value for intercept b = 0

```python
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0

loss_slope=-2 * np.sum(y-m*X.ravel()+b)
loss_slope
```

```python
#lets take learning rate 0.1
lr=0.1
step_size=loss_slope*lr
step_size
```

```python
b=b-step_size
```

```python
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),label='OLS')
plt.plot(X,y_pred,color='red',label='pred1')
plt.plot(X,y_pred2,color='green',label='pred2')
plt.legend()
plt.show()
```

```python
#Iteration 2
loss_slope= -2 * np.sum(y- m * X.ravel() -b)
```

```python
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b= b - step_size
```

```python
b
```

```python
y_pred3=((m*X)+b).reshape(4)
```

```python
y_pred3
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),label='OLS')
plt.plot(X,y_pred,label='y_pred',color='red')
plt.plot(X,y_pred2,label='y_pred2',color='green')
plt.plot(X,y_pred3,label='y_pred3',color='black')
plt.legend()
plt.show()
```

```python
#iteration 2
loss_slope= -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b = b-step_size
```

```python
b
```

```python
y_pred4=((m*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),label='OLS')
plt.plot(X,y_pred,label='y_pred',color='red')
plt.plot(X,y_pred2,label='y_pred2',color='green')
plt.plot(X,y_pred3,label='y_pred3',color='black')
plt.plot(X,y_pred4,label='y_pred4',color='pink')
plt.legend()
plt.show()
```

## When learing rate is tooo small

```python

b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
