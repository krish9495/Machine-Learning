```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import plotly.express as px
import plotly.graph_objects as go
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"D:\100 Days machine learning\multiple regression part 1 day 50\Student_Performance.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.iloc[:,0:5]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
transformer=ColumnTransformer([
    ('cat',OneHotEncoder(),[2])
],remainder='passthrough')
```

```python
X_train_transformed = transformer.fit_transform(X_train)
```

```python
X_test_transformed=transformer.transform(X_test)
```

```python
scaler=StandardScaler()
```

```python
X_train_trf=scaler.fit_transform(X_train_transformed)
X_test_trf=scaler.transform(X_test_transformed)
```

```python
lr= LinearRegression()
```

```python
lr.fit(X_train_trf,y_train)
```

```python
y_pred=lr.predict(X_test_trf)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lr, X_train_trf, y_train, cv=10, scoring='r2')
```
