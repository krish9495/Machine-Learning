```python
from sklearn.datasets import load_diabetes
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
class SGDRegressor:
    def __init__(self,learning_rate=0.1,epochs=100):
        self.coef=None
        self.intercept=None
        self.lr=learning_rate
        self.epoch=epochs

    def fit(self,X_train,y_train):
        self.intercept=0
        self.coef=np.ones(X_train.shape[1])
        for i in range(self.epoch):
            for j in range (X_train.shape[0]):
                idx=np.random.randint(0,X_train.shape[0])
                
                y_hat=np.dot(X_train[idx],self.coef)+self.intercept
                
                intercept_der = -2* (y_train[idx]-y_hat)
                self.intercept=self.intercept - (self.lr* intercept_der)

                coef_der= -2 * np.dot((y_train[idx]-y_hat),X_train[idx])
                self.coef=self.coef - (self.lr * coef_der)

        print(self.coef,self.intercept)


    def predict(self,X_test):
        return np.dot(X_test,self.coef)+ self.intercept
```

```python
sgd=SGDRegressor(learning_rate=0.1,epochs=20)
```

```python
sgd.fit(X_train,y_train)
```

```python
y_pred=sgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
